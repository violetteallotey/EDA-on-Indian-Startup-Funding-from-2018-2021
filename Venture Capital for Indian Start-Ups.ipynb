## QUESTIONS AND ANSWER

### 1. How does the age of a company impact investment opportunity?


```python
# A look at the count, total sum and avergae of amount funding per year founded
Funding_by_Year_Founded = newdf_clean.groupby(by = "Founded")["Amount($)"].agg(["count", "sum", "mean"]).sort_values(by = "sum",
                                                                                                         ascending = False)
Funding_by_Year_Founded.head(10)
```

```python
# Plotting graphs for number of startups per year and Funding amount per year founded

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
fig.tight_layout(pad=3.0)

Founded_Count=pd.DataFrame(newdf_clean['Founded'].value_counts().sort_values(ascending=False).head(5))
Founded_Count.rename(columns={'Founded':'Number of startup by Year Founded'}, inplace=True)
Founded_Total=pd.DataFrame(newdf_clean.groupby(['Founded'])['Amount($)'].sum().sort_values(ascending=False).head(5))

sns.barplot(x=Founded_Count.index, y=Founded_Count['Number of startup by Year Founded'], data=Founded_Count, ax=axes[0])
sns.barplot(x=Founded_Total.index, y=Founded_Total['Amount($)'], data=Founded_Total, ax=axes[1])

plt.show()
```

### 2. Which investment level stage attract more investors?


```python
# Plotting graphs for investment level stage attract more investors

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
fig.tight_layout(pad=3.0)

Stage_Count=pd.DataFrame(newdf_clean['Stage'].value_counts().sort_values(ascending=False).head(5))
Stage_Count.rename(columns={'Stage':'Number of startup by Stage'}, inplace=True)
Stage_Total=pd.DataFrame(newdf_clean.groupby(['Stage'])['Amount($)'].sum().sort_values(ascending=False).head(5))

sns.barplot(x=Stage_Count.index, y=Stage_Count['Number of startup by Stage'], data=Stage_Count, ax=axes[0])
sns.barplot(x=Stage_Count.index, y=Stage_Total['Amount($)'], data=Stage_Total, ax=axes[1])

plt.show()
```

##### From the analysis, Venture-Series Unknown is in the majority for funding for startups, but this is not relevant since the series of funding is unknown. Hence, it can be argued that investors provide additional money to a start-up during its seed stage, then again during its Series A stage. This demonstrates that the majority of investors in the Indian startup ecosystem favor early investments.
